### We are using this notebook to execute parts of the function to see if we are going in the right direction


```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Read data from file
df = pd.read_csv("epa-sea-level.csv")
df
```

```python
# Create scatter plot
fig, ax = plt.subplots(figsize=(15, 6))
plt.style.use("dark_background")
plt.scatter(df["Year"], df["CSIRO Adjusted Sea Level"], c="b", label="Data")
```

```python
lineA = linregress(df["Year"], df["CSIRO Adjusted Sea Level"])
x = range(1880, 2051)
y = lineA.slope * x + lineA.intercept
plt.plot(x, y, "r", label="Best Fit Line 1")
```

```python
df_2000 = df[df["Year"] >= 2000]
line_2 = linregress(df_2000["Year"], df_2000["CSIRO Adjusted Sea Level"])
x_2000 = range(2000, 2051)
y_2000 = line_2.slope * x_2000 + line_2.intercept
plt.plot(x_2000, y_2000, "g", label="Best Fit Line 2")

# Add labels and title
plt.xlabel("Year")
plt.ylabel("Sea Level (inches)")
plt.title("Rise in Sea Level")
plt.legend()
```
